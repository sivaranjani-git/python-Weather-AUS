```python
df_num=df.select_dtypes(include=['int64','float64'])
from sklearn.preprocessing import StandardScaler,MinMaxScaler
sc=StandardScaler()
df_num_sc=sc.fit_transform(df_num)
df_num_sc=pd.DataFrame(data=df_num_sc,columns=df_num.columns)
df_num_sc
```

```python
df_cat=df.select_dtypes(include=['object'])
dummy=pd.get_dummies(data=df_cat['WindGustDir'],columns=['WindGustDir'],drop_first=False)
dummy
```

```python
df_c=pd.concat([df_num_sc,dummy],axis=1)
df_c
```

```python
X = df_c
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =0)
```

```python
rfc = RandomForestClassifier(n_estimators=100,random_state=0)
```
